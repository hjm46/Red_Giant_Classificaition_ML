# META DATA EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
df.head()
```

### Graphs

```python
sns.histplot(data=df["SNR"], bins=50)
plt.title("Signal to Noise Ratio Histogram with 50 Bins")
plt.xlabel("Signal to Noise Ratio")
plt.ylabel("Occurance Count")
```

```python
sns.histplot(data=df["RV_TEFF"], bins=50)
plt.title("Effective Temperature Histogram with 50 Bins")
plt.xlabel("Effective Temperature [K]")
plt.ylabel("Occurance Counts")
```

```python
sns.histplot(data=df["RV_LOGG"], bins=50)
plt.title("Log Effective Surface Gravity Histogram with 50 Bins")
plt.xlabel("Log Effective Surface Gravity [log(m/s^2)]")
plt.ylabel("Occurance Counts")
```
