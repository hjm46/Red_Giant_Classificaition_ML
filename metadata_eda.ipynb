### Differentiating between Red Giant and Red Clump stars

Using logic from the Bovy Paper.

1.8 <= logg <= 0.0018 dex K^−1 Teff−Tref_eff ([Fe/H])+2.5,

where

Tref_eff ([Fe/H]) =−382.5K dex^−1 [Fe/H] + 4607K

```python
# using shuffled data for training
df_shuffled = pd.read_csv("cleaned_data_shuffled.csv")

t_ref = -382.5 * df_shuffled['RV_FEH'] + 4607
delta_t = df_shuffled['RV_TEFF'] - t_ref
logg_upper_bound = (0.0018 * delta_t) + 2.5

rc_mask = df_shuffled["RV_LOGG"].between(1.8, logg_upper_bound)
df_shuffled["Stellar_type"] = np.where(rc_mask, "Red Clump", "Red Giant")
```

```python
sns.scatterplot(
    data=df_shuffled, 
    x="RV_TEFF", 
    y="RV_LOGG", 
    hue="Stellar_type", 
    palette={"Red Clump": "#FFD700", "Red Giant": "#DC143C"},
    s=5, 
    alpha=0.6,
    edgecolor=None
)

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title('HR Diagram: Red Clump vs. Red Giant Branch')
plt.xlabel('Effective Temperature [K]')
plt.ylabel('Surface Gravity [log(m/s^2)]')
plt.legend(title='Stellar Classification', loc='upper left')

plt.show()
```
